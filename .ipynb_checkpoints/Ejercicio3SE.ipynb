```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
X = bd_horm[["horas"]]
y = bd_horm[["monto"]]
```

```python
#Haremos un boostrap sobre la variable de respuesta Y 
#en cada iteracion se hara un modelo lineal
#y en cada ireracion se calculara el error estandar de las betas
#y comparar con el
#1000 modelos con las y's remuestradas, con las y boostrap
```

```python
Y=y.to_numpy().reshape(-1)
Y
```

```python
#hacemos un remuestreo con reemplazo en python de la variable de respuesta Y
import pandas as pd
import numpy as np

df = pd.DataFrame(Y,
                 columns=['monto'])
df["horas"]=X
#df

r_i=np.random.choice(Y,len(Y),replace=True)
#r_i

df_aux = pd.DataFrame(r_i,
                 columns=['monto'])
#df_aux

#las muestras con reemplazo las juntamos con el data frame df, mediante el monto.
df_aux=pd.merge(df_aux, df, how="inner", on=["monto"])
df_aux
```

```python
simul=1000
res_boost= np.array( [ ])
res_boost1= np.array( [ ])
for i in range(0,simul):
    r_i=np.random.choice(Y,len(Y),replace=True)
    df_aux = pd.DataFrame(r_i,
                 columns=['monto'])
    df_aux=pd.merge(df_aux, df, how="inner", on=["monto"])
    df_aux
    #x_i=df_aux[["x"]].to_numpy().reshape(-1)
    
    xi = df_aux[["horas"]]
    yi = df_aux[["monto"]]
    #Hacemos la regresion
    reg.fit(xi, yi)
    
    #Obtenemos el intercepto B0
    B0=reg.intercept_
    #Obtenemos el coeficiente, B1
    #un coeficiente porque solo tenemos una variable explicativa
    B1=reg.coef_
    
    
    res_boost = np.append(res_boost,B0)
    res_boost1 = np.append(res_boost1,B1)
    
    #Datos estimados
    yi_pred = reg.predict(xi) 
    yi_estim=yi_pred.reshape(-1)
    
    x=xi.to_numpy().reshape(-1)
    y_real=yi.to_numpy()
    y_real=y_real.reshape(-1)
    n=len(y_real)

    
```

```python
#Beta0 es el vector de 1000 simulaciones usando Boostrap
beta0=res_boost
beta0
```

```python
beta1=res_boost1
```

```python
#Beta1 es el vector de 1000 simulaciones usando Boostrap
beta1
```

Podemos obtener el error estándar del estimador de la siguiente manera
$$
\begin{equation}
\hat{se}_{\beta}=\sqrt{\frac{\sum_{i} \epsilon_{i}^{2}}{(n-2)\sum_{i}(x-\bar{x})^{2}}}
\end{equation}
$$

```python
n=1000
#Calcularemos el error estandar de las betas 

#Para Beta0
den=sum((x-np.mean(x))**2)
num=sum(beta0**2)/(n)
err_reg=np.sqrt(num/den)
err_reg
```

```python
#Como podemos observar el error estandar para B0 es pequeño
```

```python
n=1000 #Es el numero de simulaciones
#Calcularemos el error estandar de las betas 

#Para Beta1
den=sum((x-np.mean(x))**2)
num=sum(beta1**2)/(n)
err_reg2=np.sqrt(num/den)
err_reg2
```

```python
#Como podemos observar el error estandar para B1 es muy pequeño
```
