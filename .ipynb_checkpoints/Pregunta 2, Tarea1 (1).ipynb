 De la base data_schools sacar obtener el error estándar con Bootstrap de las variables de la base. Recordemos que tienen que simular $N$ Muestras Boostrap con tamño $n$. Luego, el otro ejercicio es que lo hagan con $n-1$.

```python
from pylab import*
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats 
from scipy.special import gamma
import seaborn as sns
from scipy.stats import beta
from scipy.stats import spearmanr
import pandas as pd
```

```python
bd_escuelas=pd.read_excel("C:/Users/AngelJoab/Documents/data_schools.xlsx")

bd_escuelas
```

```python
bd_escuelas = bd_escuelas[["LSAT","GPA"]]
bd_escuelas 
```

```python
# Desviación estandar de LSAT y GPA
np.std(bd_escuelas)
```

```python
#Medias de LSAT y GPA
np.mean(bd_escuelas)
```

```python
#Varianza de LSAT y GPA
np.var(bd_escuelas)
```

```python
corr = spearmanr(bd_escuelas.LSAT, bd_escuelas.GPA)[0]
print("La correlacion de la muestra es:", corr)
```

Definimos una funcion que nos genere una muestra de tamano $n$ de la correlacion entre LSAT y GPA

```python
def Generarmuestra(x, y):
    muestra = list()
    for i in range(x):
        muestra_i = bd_escuelas.sample(y)
        muestra.append(spearmanr(muestra_i))
    return muestra

```

```python
#Obtenemos la siguiente lista de correlaciones de 10,000 simulaciones de tamaño 15
m = Generarmuestra(x=10000, y=15)
m 
```

```python
dm = pd.DataFrame(m)
dm
```

```python
dm = dm[['correlation']]
dm
```

```python
#Se generan las 10,000 muestras bootstrap sobre la correlación de Spearman entre LSAT y GPA
boot_corrs = []

bootsample_corr = dm
boot_corrs.append(dm)
boot_corrs_np = np.array(boot_corrs)
boot_corrs_np
```

```python
np.mean(boot_corrs_np)
print("El promedio de las correlaciones de las muestras bootstrap es:",np.mean(boot_corrs_np))
```

```python
corr = spearmanr(bd_escuelas.LSAT, bd_escuelas.GPA)[0]
print("La correlacion de la muestra total es:", corr)
```

```python
np.std(boot_corrs_np)  #el error estandar con Bootstrap de la correlación. 
```

```python
np.percentile(boot_corrs_np, [1.5,98.5]) #es el 97% intervalo de confianza
```
